## Fetch Verification Code from Email

```python
def get_verification_code():
    with IMAPClient('imap.gmail.com') as server:
        server.login(EMAIL, APP_PASSWORD)
        server.select_folder('INBOX', readonly=True)

        messages = server.search(['UNSEEN'])

        for uid in reversed(messages):
            raw = server.fetch([uid], ['BODY[]'])
            msg = pyzmail.PyzMessage.factory(raw[uid][b'BODY[]'])

            if msg.text_part:
                body = msg.text_part.get_payload().decode(msg.text_part.charset)

                match = re.search(r"\\b\\d{6}\\b", body)
                if match:
                    return match.group(0)

    return None

code = None
for _ in range(10):
    code = get_verification_code()
    if code:
        break
    time.sleep(5)

print("Code:", code)
```

```python
import os.path
from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build

# If modifying these scopes, delete the file token.json.
SCOPES = ['https://www.googleapis.com/auth/gmail.readonly']

def get_latest_email():
    creds = None
    # The file token.json stores the user's access and refresh tokens
    if os.path.exists('token.json'):
        creds = Credentials.from_authorized_user_file('token.json', SCOPES)
    
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file('credentials.json', SCOPES)
            creds = flow.run_local_server(port=0)
        with open('token.json', 'w') as token:
            token.write(creds.to_json())

    service = build('gmail', 'v1', credentials=creds)

    # Call the Gmail API to list messages from Inbox
    results = service.users().messages().list(userId='me', labelIds=['INBOX'], maxResults=1).execute()
    messages = results.get('messages', [])

    if not messages:
        print("No messages found.")
    else:
        # Get the details of the most recent message
        message = service.users().messages().get(userId='me', id=messages[0]['id']).execute()
        
        print(f"Snippet: {message['snippet']}")
        
        # To get the Subject, loop through headers
        headers = message['payload']['headers']
        subject = next(h['value'] for h in headers if h['name'] == 'Subject')
        sender = next(h['value'] for h in headers if h['name'] == 'From')
        
        print(f"From: {sender}")
        print(f"Subject: {subject}")

if __name__ == '__main__':
    get_latest_email()
```

```python
import imaplib
import email
from email.header import decode_header

# --- Configuration ---
EMAIL_USER = "arisobrooks@gmail.com"
# Enter your 16-character App Password here (no spaces)
EMAIL_PASS = "wtmc fofd lptd hepp" 

def get_last_email():
    try:
        # Connect to Gmail's IMAP server
        mail = imaplib.IMAP4_SSL("imap.gmail.com")
        mail.login(EMAIL_USER, EMAIL_PASS)
        
        # Select the folder you want to check
        mail.select("inbox")

        # Search for all emails in the inbox
        # status, data = mail.search(None, "ALL") # Use this for all
        status, data = mail.search(None, 'ALL')
        
        # Get the list of email IDs and pick the last one
        mail_ids = data[0].split()
        latest_email_id = mail_ids[-1]

        # Fetch the email data (RFC822 is the standard format)
        status, data = mail.fetch(latest_email_id, "(RFC822)")

        for response_part in data:
            if isinstance(response_part, tuple):
                # Parse the raw bytes into a Python email object
                msg = email.message_from_bytes(response_part[1])
                
                # Decode the email subject
                subject, encoding = decode_header(msg["Subject"])[0]
                if isinstance(subject, bytes):
                    subject = subject.decode(encoding if encoding else "utf-8")
                
                # Get the Sender
                from_ = msg.get("From")

                print(f"--- Latest Email Found ---")
                print(f"From: {from_}")
                print(f"Subject: {subject}")
                
        # Close the connection
        mail.close()
        mail.logout()

    except Exception as e:
        print(f"Error: {str(e)}")

if __name__ == "__main__":
    get_last_email()
```
